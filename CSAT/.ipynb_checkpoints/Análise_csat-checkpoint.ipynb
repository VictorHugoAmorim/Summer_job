### <center> Gerar gráficos de notas </center>

Quantidade de pessoas que votaram de 1 a 4

```python
def gerar_grafico_seller(data,title):
    group = data.groupby('q1_opcao').count()
    group.reset_index(inplace= True)
    group = group[['q1_opcao','q1_nota']]
    percent = list(round(data.q1_nota.value_counts(normalize = True),2))
    group.insert(2,'percentual',percent)
    #graph
    fig = px.bar(group,x ='q1_opcao' , y = 'q1_nota', color = 'q1_opcao',title = title)
    return print(group),fig.show()
```

```python
def gerar_grafico_buyer(data):
    group = data.groupby('q1_opcao').count()
    group.reset_index(inplace= True)
    group = group[['q1_opcao','q1_nota']]
    percent = list(round(data.q1_nota.value_counts(normalize = True),2))
    group.insert(2,'percentual',percent)
    # Graph
    my_circle = plt.Circle( (0,0),0.7, color= 'white')
    fig = plt.figure()
    plt.pie(group.q1_nota, labels = group.q1_opcao)
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    print(group)
    return plt.show()
```

## <center> `buyer_content_discovery` </center>


```python
gerar_grafico_buyer(df_buyer_content_discovery)
```

## <center> `buyer_negotiation` </center>
